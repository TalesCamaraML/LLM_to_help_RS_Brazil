```python
###################### Carregando e configurando o modelo ##########

import google.generativeai as genai

from google.colab import userdata

CAMARA_API_KEY = userdata.get('CAMARA_API_KEY')

genai.configure(api_key=CAMARA_API_KEY)

model = genai.GenerativeModel('gemini-1.5-pro-latest')

####################################################################
import ipywidgets as widgets
from IPython.display import display, HTML, clear_output

import pathlib
import textwrap

from IPython.display import display
from IPython.display import Markdown

def to_markdown(text):
  text = text.replace('‚Ä¢', '  *')
  return Markdown(textwrap.indent(text, '> ', predicate=lambda _: True))

chat = model.start_chat(history=[])

prompt = input('Como posso te ajudar? \n')

while prompt != "parar":
  clear_output()
  print("Gerando resposta...")  # Exibe a mensagem
  resposta = chat.send_message(prompt)
  display( to_markdown(resposta.text) )
  print("-----------------------------------------------------------")
  prompt = input('Gostaria de mais alguma coisa? \n')
```

# **Identifique Vulnerabilidades ap√≥s Trag√©dias Clim√°ticas com Nossa Aplica√ß√£o**

---



Ap√≥s uma trag√©dia clim√°tica, pode ser desafiador identificar indiv√≠duos e animais em situa√ß√£o de vulnerabilidade. Para auxiliar nesta tarefa delicada, desenvolvemos uma aplica√ß√£o que permite que voc√™ carregue imagens de seu dispositivo.

A aplica√ß√£o fornece uma descri√ß√£o abrangente do cen√°rio, destacando √°reas de risco e recursos dispon√≠veis. Al√©m disso, ela usa tecnologia avan√ßada para identificar automaticamente indiv√≠duos que precisam de ajuda ou est√£o em perigo.

Ao usar esta aplica√ß√£o, voc√™ pode contribuir significativamente para os esfor√ßos de socorro, garantindo que os necessitados recebam assist√™ncia o mais r√°pido poss√≠vel.

**Como Usar:**

1. Carregue uma imagem do cen√°rio afetado.
2. A aplica√ß√£o fornecer√° uma descri√ß√£o geral do cen√°rio, incluindo √°reas de risco e recursos dispon√≠veis.
3. A aplica√ß√£o tamb√©m identificar√° automaticamente indiv√≠duos que precisam de ajuda ou est√£o em perigo.
4. Compartilhe as informa√ß√µes com as equipes de resposta a emerg√™ncias para coordenar os esfor√ßos de socorro.

**Benef√≠cios:**

* Identifica√ß√£o r√°pida e precisa de indiv√≠duos vulner√°veis
* Descri√ß√£o abrangente do cen√°rio, incluindo √°reas de risco e recursos

**Observa√ß√£o:**

Esta aplica√ß√£o ainda est√° em desenvolvimento. Agradecemos quaisquer coment√°rios ou sugest√µes que voc√™ possa ter.

```python
from pathlib import Path
import io
import google.generativeai as genai
from google.colab import files
import matplotlib.pyplot as plt

from IPython.display import display
from IPython.display import Markdown

def to_markdown(text):
  text = text.replace('‚Ä¢', '  *')
  return Markdown(textwrap.indent(text, '> ', predicate=lambda _: True))

system_instruction = "Voc√™ √© um especialista em analise de imagens.Em portugu√™s, fa√ßa uma descri√ß√£o geral da imagem, destaque √°reas de risco, retorne quantos animais (humanos e n√£o humanos) existem nas imagens. A resposta deve ser emp√°tica"

model = genai.GenerativeModel(model_name="gemini-1.5-pro-latest",
                              system_instruction=system_instruction)

def uploadfile():
  print("Carregue suas imagens")
  uploaded = files.upload()
  filesload = []
  arquivos_txt = {nome: conteudo for nome, conteudo in uploaded.items() if (nome.endswith('.jpg') or nome.endswith('.jepg') or nome.endswith('.png')) }
  nomes_imagens = []

  for nome, conteudofile in arquivos_txt.items():
    print(f"Nome da imagem: {nome}")
    filesload.append(genai.upload_file(f"/content/{nome}" ))
    nomes_imagens.append(f"/content/{nome}")

  for nome_imagem in nomes_imagens:
    imagem = plt.imread(nome_imagem)
    plt.imshow(imagem)
    plt.show()

  return filesload

def loadPromt():
  filesload = uploadfile()
  prompt_parts = filesload

  resposta = model.generate_content(prompt_parts)
  print("---------------------------------------------------")

  display( to_markdown(resposta.text) )

  print("---------------------------------------------------")

loadPromt()

```
